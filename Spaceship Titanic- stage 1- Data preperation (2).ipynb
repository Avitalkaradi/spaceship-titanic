Changing Object data type so we can manipulate the data and merge the tables

```python
df1['PassengerId'] = df1['PassengerId'].str.replace('_', '').astype(int)
df2['PassengerId'] = df2['PassengerId'].str.replace('_', '').astype(int)
```

# Strings

```python
#Name
df1['Name'] = df1['Name'].astype('string')
df2['Name'] = df2['Name'].astype('string')
#HomePlanet    
df1['HomePlanet'] = df1['HomePlanet'].astype('string')
df2['HomePlanet'] = df2['HomePlanet'].astype('string')
#Destination
df1['Destination'] = df1['Destination'].str.replace('-1e', '').astype('string')
df2['Destination'] = df2['Destination'].str.replace('-1e', '').astype('string')
df1['Destination'] = df1['Destination'].str.replace('J318.5-22', '').astype('string')
df2['Destination'] = df2['Destination'].str.replace('J318.5-22', '').astype('string')
df1['Destination'] = df1['Destination'].astype('string')
df2['Destination'] = df2['Destination'].astype('string')
```

# bool

```python
df1['CryoSleep'] = df1['CryoSleep'].replace(to_replace = "True", value = 1).astype(float)
df1['CryoSleep'] = df1['CryoSleep'].replace(to_replace = "False", value = 0).astype(float)
df2['CryoSleep'] = df2['CryoSleep'].replace(to_replace = "True", value = 1).astype(float)
df2['CryoSleep']= df2['CryoSleep'].replace(to_replace = "False", value = 0).astype(float)
```

```python
df1['VIP'] = df1['VIP'].replace(to_replace = "True", value = 1).astype(float)
df1['VIP'] = df1['VIP'].replace(to_replace = "False", value = 0).astype(float)
df2['VIP'] = df2['VIP'].replace(to_replace = "True", value = 1).astype(float)
df2['VIP']= df2['VIP'].replace(to_replace = "False", value = 0).astype(float)
```

```python
# Split the column into three columns
df1[['a', 'b', 'c']] = df1['Cabin'].str.split('/', expand=True).astype('string')
df2[['a', 'b', 'c']] = df2['Cabin'].str.split('/', expand=True).astype('string')
```

```python
df1['Cabin'] = df1['Cabin'].str.replace('/', '').astype('string')
df2['Cabin'] = df2['Cabin'].str.replace('/', '').astype('string')
```

```python
df2['Transported'] =df2['Transported'].astype('string')
df1['CryoSleep'] = df1['CryoSleep'].astype('string')
df2['CryoSleep'] = df2['CryoSleep'].astype('string')
df1['VIP'] = df1['VIP'].astype('string')
df2['VIP'] = df2['VIP'].astype('string')
df1['PassengerId'] = df1['PassengerId'].astype('string')
df2['PassengerId'] = df2['PassengerId'].astype('string')
```

```python
merged_df = pd.concat([df1, df2], axis=0)
merged_df.reset_index(drop=True, inplace=True)
merged_df.shape
```
